```python
from lifelines import CoxPHFitter
import pandas as pd
```

```python
surv_df = pd.read_csv('data_clean_survival.csv')
```

```python
surv_df.head()
```

```python
surv_df = surv_df.drop(['Unnamed: 0','Close.Date','First.Deal.Created.Date',
                       'Last.Call.Cycle.Date','Time.of.Last.Session'], axis = 1)
```

```python
surv_df.head()
```

```python
surv_df = surv_df.dropna()
```

```python
#Cox Proportional Hazards Model
cph = CoxPHFitter(penalizer=0.001)   ## Instantiate the class to create a cph object
cph.fit(surv_df, 'days_live', event_col='churn')   ## Fit the data to train the model
cph.print_summary()    ## HAve a look at the significance of the features
```

```python
cph.plot()
```

### Customer Survival Curves

```python
cust_select = surv_df.iloc[5:10, 1:]
cust_select
```

```python
#Predicted Survival Curves of Selected Customers
cph.predict_survival_function(cust_select).plot()
```

```python
surv_vis = pd.read_csv('survival_vis.csv')
surv_vis.head()
```

```python
cph.predict_survival_function(surv_vis).plot().axis([0, 1000, 0.9, 1.0])
```
